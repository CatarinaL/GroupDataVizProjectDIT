## Exploratory Data Analysis

### 1. Reading the data

Read and convert the data into Python dictionaries

```python
import csv 

file_path = "dataset/clean_v1_Airplane_Crashes_and_Fatalities_Since_1908.csv"

with open(file_path) as file:
    reader = csv.DictReader(file) #row iterator
    crashes = list(reader)
    
crashes
```

Put the data into a Pandas dataframe:



```python
import pandas as pd

dframe = pd.DataFrame(crashes)
dframe

```

Alternatively, the two previous blocks can be simply handled by Pandas method "read_csv", as below:

```python
import pandas as pd

dframe = pd.read_csv(file_path)

# get information about data entries by column:
dframe.info()
# get a brief statistical summary of the data
dframe.describe()
```
